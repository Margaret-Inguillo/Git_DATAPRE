```python
import requests
import json

URL="https://www.rappler.com/united-states"
```

```python
page=requests.get(URL)
```

```python
print(page.content)
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# Get main content <div> class
art_content=soup.find(class_="StyledComponents__ItemWrapper-hzks6l-1 dCoFec")
#print(art_content)
# Get all <div> elements
#Get the contents under the main content of <div> class attribute of each article
art_p=art_content.find_all('p')
print(art_p)
art_h3=art_content.find_all('h3')
print(art_h3)
art_time=art_content.find_all('time')
print(art_time)
```

```python
for i in range(len(art_h3)):
    
    print(f'Index {i} - {art_h3[i]}\n\n')
```

```python
for i in range(len(art_h3)):
    #art_h3[i].text.strip()
    #art_p[i].text.strip()
    #art_time[i].text.strip()
    print(f'Index {i} - {art_h3[i].text.strip()}')
    print(f'Category - {art_p[i].text.strip()}')
    print(f'Time - {art_time[i].text.strip()}\n\n')
```

```python
gen1_json = [] #create a blank list of json file
#printing the for loop to store the values
#------------------
#storing of indv values 
#adding json object/s
for i in range(len(art_h3)):
        sava=art_h3[i].text.strip()
        savp=art_p[i].text.strip()
        savt=art_time[i].text.strip()   
        gen1_json.append({
            
            "Title": sava,
            "Category": savp,
            "Time": savt 
        })
                
with open('Rappler_US.json', 'w') as json_file:
     json.dump(gen1_json, json_file)    
            
gen1_json
```
